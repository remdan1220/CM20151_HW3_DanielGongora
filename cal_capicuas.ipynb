```python
n = raw_input("texto >")

m = n.replace(' ','').replace(',','').replace('!','').replace(':','').replace(';','').replace('.','').replace('-','').lower()
f = m.decode('unicode-escape')

import unicodedata
```

```python
def tildes_out(s):

    nk = unicodedata.normalize('NFKD', s)
    only = nk.encode('ASCII', 'ignore')
    return only

ng = tildes_out(f)
nf = ng.lower()
if nf==nf[::-1]:
    print "el texto si es palindromo"
else:
    print "el texto No es palindromo"
print nf
```
